## Step 4: Configure Chat Interface And Try Out Your Agent

Now that our frontend chat interface and our backend agent are both running, let's try chatting with our agent!

First, you must open up your agent to traffic so you can chat with it through your browser. To do that, navigate to your launchable's instance page on brev and scroll down to the **Using Secure Links** section. Find the service that's on port `2024` and click **Edit Access**. Then, click the toggle next to **Make Public**. Now you're agent is ready to receive messages!

To open up the chat interface, look at the rest of the URLs under that **Using Secure Links**. You should see another service on port `3000`. You can click that link and it will open up a chat interface in your browser. When it opens, you will see a screen asking you to set up a configuration:

For the **Deployment URL**, go back to the "Using Secure Links" section on Brev and find the link that's on port `2024`. Copy it and paste it here as the Deployment URL.

For the **Assitant ID**, put `agent`.
